```python
import open3d as o3d
import requests

# URL for the point cloud file
url = 'https://raw.githubusercontent.com/PointCloudLibrary/pcl/master/test/bunny.pcd'

# Download the file locally
response = requests.get(url)
filename = 'bunny.pcd'
with open(filename, 'wb') as f:
    f.write(response.content)

# Load the point cloud from the downloaded file
pcd = o3d.io.read_point_cloud(filename)

# Check if the point cloud object was successfully loaded
if pcd is not None:
    print("Point cloud loaded successfully.")
    print(pcd)
else:
    print("Failed to load the point cloud.")
```
